# Acá debemos decidir cuáles especialidades ('serv_codigo' y 'serv_nombre') y entender mejor las columnas 'clpr_nombre' y 'clpr_codigo' para decidir si dejamos las ips, profesionales independientes...

```python
medicos_ips = medicos_ips[medicos_ips.serv_codigo == 328]
reemplazo = {'San Andrés y Providencia':'Archipiélago de San Andrés', 'Valle del cauca':'Valle del Cauca', 'Bogotá D.C':'Bogotá, D.C.', 'Quindío':'Quindio'}
medicos_ips.depa_nombre.replace(reemplazo, inplace = True)
EDA2020 = medicos_ips['depa_nombre'].value_counts().sort_index()
EDA2020
```

```python
pob = pd.read_parquet(f'{dir_data}/proyecciones_poblacion_departamental_2018-2050.parquet')
pob
```

```python
pob = pob[pob['ÁREA GEOGRÁFICA'] == 'Total']
pob = pob[pob['AÑO'] == 2020]
pob = pob.loc[:, ['DP','DPNOM','AÑO','ÁREA GEOGRÁFICA', 'Total']]
pob
```

```python
print(EDA2020.index == pob.DPNOM.sort_values().values)
```

```python
med_pob_2020 = pob.merge(pd.DataFrame(EDA2020), left_on = 'DPNOM', right_index = True)
med_pob_2020['Habs_por_med'] = round(med_pob_2020['Total']/med_pob_2020['depa_nombre'],2)
med_pob_2020.sort_values(by = 'Habs_por_med', ascending = False, inplace = True)
med_pob_2020
```

```python
med_pob_2020.plot.bar(x = 'DPNOM', y = 'Habs_por_med')
```
