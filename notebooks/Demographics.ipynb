## 2017 estimate

```python
dictsex={'Hombres':'Masculino','Mujeres':'Femenino'}
df18_3=df18_3.assign(Sexo = lambda x: x.Sexo.map(dictsex))
df18_3
```

```python
df18_3.isna().sum()
```

```python
#merge
df18_3=df18_3.rename(columns={'Sexo':'SEXO','DP':'DPTO'})
n18_2=n18_2.rename(columns={'COD_DPTO':'DPTO'})
df17=df18_3.merge(d18_2,how='outer').merge(n18_2, how='outer')
df17
```

```python
df17.isna().sum()
```

```python
df17.fillna(0,inplace=True)
df17['Poblacion1']=df17.Número - df17.Muertes + df17.Nacimientos
df17['Prop escala']=round(df17['Número']*1/5)
df17=df17.sort_values(by='Rango2', ascending=False).reset_index().drop(columns='index')
df17
```

```python
df17['Cambio']=df17[['DPTO','SEXO','Prop escala']].groupby(['DPTO','SEXO']).shift().fillna(0)
df17['Poblacion']=df17.Número + df17.Muertes - df17.Nacimientos-df17['Prop escala']+df17.Cambio
df17
```
